## DOUBLY LINKEDLIST

```python
class DNode:
    def __init__(self, NewData=None, plink=None, nlink=None):
        self.data = NewData
        self.previous = plink
        self.next = nlink


class DoublyLinkedList:
    def __init__(self):
        self.Head = None

    def insert_end(self, newData):
        newNode = DNode(newData)
        if self.Head is None:
            self.Head = newNode
        else:
            curNode = self.Head
            while curNode.next is not None:
                curNode = curNode.next
            curNode.next = newNode
            newNode.previous = curNode
        return "Node inserted at the end successfully."

    def insert_begin(self, newData):
        newNode = DNode(newData, None, self.Head)
        if self.Head is not None:
            self.Head.previous = newNode
        self.Head = newNode
        return "Node inserted at the beginning successfully."

    def insert_before_nth(self, newData, location):
        if location <= 0:
            return "Invalid position. Position must be greater than 0."

        newNode = DNode(newData)
        if self.Head is None or location == 1:
            return self.insert_begin(newData)

        curNode = self.Head
        c = 1
        while curNode is not None and c < location - 1:
            curNode = curNode.next
            c += 1

        if curNode is None:
            return "Position out of bounds."

        newNode.previous = curNode
        newNode.next = curNode.next
        if curNode.next is not None:
            curNode.next.previous = newNode
        curNode.next = newNode
        return "Node inserted before the specified position successfully."

    def insert_after_nth(self, newData, location):
        if location <= 0:
            return "Invalid position. Position must be greater than 0."

        newNode = DNode(newData)
        curNode = self.Head
        c = 1
        while curNode is not None and c < location:
            curNode = curNode.next
            c += 1

        if curNode is None:
            return "Position out of bounds."

        newNode.previous = curNode
        newNode.next = curNode.next
        if curNode.next is not None:
            curNode.next.previous = newNode
        curNode.next = newNode
        return "Node inserted after the specified position successfully."

    def delete_first(self):
        if self.Head is None:
            return "Empty List. Deletion not possible."
        curNode = self.Head
        self.Head = self.Head.next
        if self.Head is not None:
            self.Head.previous = None
        del curNode
        return "Node Deleted Successfully."

    def delete_last(self):
        if self.Head is None:
            return "Empty List. Deletion not possible."
        elif self.Head.next is None:
            del self.Head
            self.Head = None
            return "Node Deleted Successfully."
        else:
            curNode = self.Head
            while curNode.next.next is not None:
                curNode = curNode.next
            del curNode.next
            curNode.next = None
            return "Node Deleted Successfully."

    def delete_nth(self, posn):
        if posn <= 0:
            return "Invalid position. Position must be greater than 0."

        if self.Head is None:
            return "Empty List. Deletion not possible."
        
        curNode = self.Head
        if posn == 1:
            return self.delete_first()

        c = 1
        while curNode is not None and c < posn:
            curNode = curNode.next
            c += 1

        if curNode is None:
            return "Position out of bounds."

        if curNode.next is not None:
            curNode.next.previous = curNode.previous
        if curNode.previous is not None:
            curNode.previous.next = curNode.next

        del curNode
        return "Node Deleted Successfully."

    def display(self):
        if self.Head is None:
            return "Empty List."
        curNode = self.Head
        result = []
        while curNode is not None:
            result.append(str(curNode.data))
            curNode = curNode.next
        return " <--> ".join(result)

    def rev_display(self):
        if self.Head is None:
            return "Empty List."
        curNode = self.Head
        while curNode.next is not None:
            curNode = curNode.next
        result = []
        while curNode is not None:
            result.append(str(curNode.data))
            curNode = curNode.previous
        return " <--> ".join(result)


head = DoublyLinkedList()
while True:
    print("\n Program to implement Doubly Linked List")
    print("====================================================")
    print("1. Create/Appending the list")
    print("2. Insert Node at beginning")
    print("3. Insert Node before Nth Position")
    print("4. Insert Node after Nth node")
    print("5. Delete First Node")
    print("6. Delete Last Node")
    print("7. Delete Nth node")
    print("8. Display the list")
    print("9. Display the list in reverse order")
    print("10. Exit")
    print("===================================================")
    
    choice = input("Enter your choice: ")
    
    if choice.isdigit():
        choice = int(choice)
    else:
        print("Invalid input! Please enter a number.")
        continue

    if choice == 1:
        opt = 'Y'
        while opt.upper() == 'Y':
            num = input("Enter the data: ")
            if num.isdigit():
                print(head.insert_end(int(num)))
            else:
                print("Invalid input! Please enter a number.")
            opt = input("Enter more (Y/N): ")
    elif choice == 2:
        num = input("Enter the data: ")
        if num.isdigit():
            print(head.insert_begin(int(num)))
        else:
            print("Invalid input! Please enter a number.")
    elif choice == 3:
        loc = input("Enter the node number before which new node will be inserted: ")
        num = input("Enter the data: ")
        if loc.isdigit() and num.isdigit():
            print(head.insert_before_nth(int(num), int(loc)))
        else:
            print("Invalid input! Please enter numbers.")
    elif choice == 4:
        loc = input("Enter the Node number after which new node will be inserted: ")
        num = input("Enter the data: ")
        if loc.isdigit() and num.isdigit():
            print(head.insert_after_nth(int(num), int(loc)))
        else:
            print("Invalid input! Please enter numbers.")
    elif choice == 5:
        print(head.delete_first())
    elif choice == 6:
        print(head.delete_last())
    elif choice == 7:
        num = input("Enter the node number you want to delete: ")
        if num.isdigit():
            print(head.delete_nth(int(num)))
        else:
            print("Invalid input! Please enter a number.")
    elif choice == 8:
        print("List:", head.display())
    elif choice == 9:
        print("Reverse List:", head.rev_display())
    elif choice == 10:
        print("\nQuitting.....")
        break
    else:
        print("Invalid choice. Please enter the correct choice.")
```
